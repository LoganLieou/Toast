```python
# our features and labels should be split up as follows
# labels are df['stroke']
# features are all other values the data is fairly clean
# note that df has been applied dropna() and onehot-encoding dummies()
df = df.dropna()
df = pd.get_dummies(df)
fets = df.drop(labels=['stroke'], axis=1)
lbls = df['stroke']
```
